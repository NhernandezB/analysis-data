```python
import os
os.getcwd()
import pandas as pd
import matplotlib.pyplot as plt
# Cargar el conjunto de datos desde un archivo CSV
dataset = pd.read_csv("datasets\\forbes_best_employers_2023.csv")
```

```python

# Obtener los conteos de los valores en la columna "Nombre de la Columna"
conteos = dataset["country_territory"].value_counts()

# Crear una tabla de frecuencia
tabla_frecuencia = pd.DataFrame({'Valor': conteos.index, 'Frecuencia': conteos.values})

# Imprimir la tabla de frecuencia
print(tabla_frecuencia)

# Crear un gráfico de barras
plt.figure(figsize=(20, 10))  # Ajusta el tamaño del gráfico si es necesario
plt.bar(tabla_frecuencia['Valor'], tabla_frecuencia['Frecuencia'])
plt.text('United States', 2,  f' {200}', ha='center', va='bottom', fontsize=12, color='black')
plt.xlabel("Country")
plt.ylabel("Number of companies")
plt.title("Forbes best employers 2023")
plt.xticks(rotation=45)  # Rota las etiquetas del eje x para una mejor legibilidad
plt.show()
```

```python

# Contar cuántas veces se repite un valor específico en la columna "Nombre de la Columna"
valor_a_contar = "United States"
conteo = dataset["country_territory"].value_counts()[valor_a_contar]

print(f"El valor '{valor_a_contar}' se repite {conteo} veces en la columna 'Nombre de la Columna'.")
```

```python
# Obtener los conteos de los valores en la columna "Nombre de la Columna"
conteos = dataset["industries"].value_counts()

# Crear una tabla de frecuencia
tabla_frecuencia = pd.DataFrame({'Industry': conteos.index, 'Frecuency': conteos.values})

# Imprimir la tabla de frecuencia
print(tabla_frecuencia)

# Crear un gráfico de barras
plt.figure(figsize=(20, 10))  # Ajusta el tamaño del gráfico si es necesario
plt.bar(tabla_frecuencia['Industry'], tabla_frecuencia['Frecuency'])
plt.xlabel("Country")
plt.ylabel("Number of companies")
plt.title("Forbes best employers 2023")
plt.xticks(rotation=90)  # Rota las etiquetas del eje x para una mejor legibilidad
plt.show()
```

```python
import matplotlib.pyplot as plt

# Crear un gráfico de dispersión
plt.figure(figsize=(20, 10))  # Ajusta el tamaño del gráfico si es necesario
plt.scatter(dataset["country_territory"], dataset["industries"], cmap='viridis')
plt.xlabel("Country")
plt.xticks(rotation=90)  
plt.ylabel("Industry")

# Guardar la imagen en un archivo
plt.savefig("grafico_relacion.png")

plt.show()
```
